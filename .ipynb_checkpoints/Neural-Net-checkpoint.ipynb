```python
%matplotlib inline

import numpy as np

from sklearn import linear_model, datasets, neural_network
from sklearn.preprocessing import OneHotEncoder, scale
import mnist_loader
train_data, validation_data, test_data = mnist_loader.load_data_wrapper()

train_X, train_Y = zip(*train_data)
cv_X, cv_Y = zip(*validation_data)
test_X, test_Y = zip(*test_data)

train_X = np.squeeze(np.asarray(train_X))
train_Y = np.squeeze(np.asarray(train_Y))

cv_X = np.squeeze(np.asarray(cv_X))
cv_Y = np.squeeze(np.asarray(cv_Y))

test_X = np.squeeze(np.asarray(test_X))
test_Y = np.squeeze(np.asarray(test_Y))
```

```python
from NN import NN

nn1 = NN([784, 50, 10], activation="softplus")
nn1.train(train_X, train_Y, 100, 0.05, 9, record_loss=True, test_data=(cv_X, cv_Y))
nn1.show_loss()
print(nn1.acc(test_X, test_Y))
```

```python
nn1 = NN([784, 30, 10], activation="tanh")
nn1.train(train_X, train_Y, 100, 0.2, 12, record_loss=True, test_data=(cv_X, cv_Y))
nn1.show_loss()
print(nn1.acc(test_X, test_Y))
```
